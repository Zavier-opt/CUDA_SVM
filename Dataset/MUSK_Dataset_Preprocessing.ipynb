```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
# Import musk dataset
MUSK = pd.read_csv("MUSK_original.csv", header = None)
MUSK.head(10)
```

```python
# Drop the first two columns
MUSK = MUSK.iloc[:,2:]
MUSK.columns = range(MUSK.shape[1])
print(MUSK.shape)
MUSK.head(10)
```

```python
# Reserve 70% of dataset for training 30% for testing
ratio = 0.7
index = int(MUSK.shape[0]*ratio)
MUSK_training = MUSK.iloc[:index,:]
MUSK_testing = MUSK.iloc[index:,:]
print(MUSK_training.shape)
print(MUSK_testing.shape)
```

```python
# Generalize the data with scale of training dataset
scaler = StandardScaler()
MUSK_training_X = scaler.fit_transform(MUSK_training.iloc[:,:-1])
MUSK_training_y = MUSK_training.iloc[:,-1]
MUSK_testing_X = scaler.transform(MUSK_testing.iloc[:,:-1])
MUSK_testing_y = MUSK_testing.iloc[:,-1]
```

```python
# Concate the scaled data with label(y)
MUSK_training_X = pd.DataFrame(MUSK_training_X)
MUSK_testing_X = pd.DataFrame(MUSK_testing_y)
MUSK_training = pd.concat([MUSK_training_X, MUSK_training_y], axis = 1)
MUSK_testing = pd.concat([MUSK_testing_X, MUSK_testing_y], axis = 1)
```

```python
MUSK_training.to_csv("MUSK_training.csv", index = True)
MUSK_testing.to_csv("MUSK_testing.csv", index = True)
```
